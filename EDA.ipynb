```python
# Convert 'Has Online delivery' and 'Has Table booking' to numeric if they are not already
df['Has Online delivery'] = df['Has Online delivery'].apply(lambda x: 1 if x == 'Yes' else 0)
df['Has Table booking'] = df['Has Table booking'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
df5 = df[['Average Cost for two', 'Has Table booking', 'Has Online delivery']]
```

```python
df5
```

```python
df['Average Cost for two'].describe().to_frame
```

```python
custom_bins = [0, 200, 400, 600, 1000, 800000]
bin_labels = ['0-200', '200-400', '400-600', '600-1000', '1000+']
legends=df[['Has Table booking','Has Online delivery']]

# Create a new column with binned data
df['Cost Range'] = pd.cut(df['Average Cost for two'], bins=custom_bins, labels=bin_labels)
```

```python
# Plot the count of observations in each bin
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Cost Range', palette="viridis", edgecolor='black',hue='Has Online delivery')
```

### Cost Range vs. Online Delivery Availability

1. **0-200**: Predominantly lacks online delivery; low-cost restaurants may not prioritize this service.
2. **200-600**: Limited online delivery, though slightly more common than in the lowest range.
3. **600+**: Higher-cost restaurants are more likely to offer online delivery, with an increase in the 1000+ range.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Cost Range', palette="viridis", edgecolor='black',hue='Has Table booking')
```

### Cost Range vs. Table Booking Availability

1. **0-200**: Minimal table booking; budget-friendly spots typically operate without reservations.
2. **200-600**: Rarely offer table booking, indicating a preference for walk-ins at mid-cost levels.
3. **600+**: Significant increase in table booking for 1000+ range, reflecting premium customer expectations.
